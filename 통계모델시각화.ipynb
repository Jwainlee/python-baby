```python
import matplotlib.pyplot as plt
import seaborn as sn

iris = sn.load_dataset('iris')
flights = sn.load_dataset('flights')
```

```python
flights.head()
```

```python
sn.lineplot(x = 'year', y = 'passengers', data = flights)
plt.show()
```

```python
sn.lineplot(x = 'year', y = 'passengers', hue = 'month', data = flights)
plt.show()
```

```python
sn.relplot(x = 'year', y = 'passengers', data = flights)
plt.show()
```

```python
sn.regplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

```python
sn.lmplot(x = 'sepal_length', y = 'petal_length', data = iris, ci = 95)
plt.show()
```

```python
sn.lmplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
plt.show()
```
