```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
iris = sn.load_dataset('iris')
iris.head()
```

```python
sn.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

```python
plt.scatter(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

```python
species = []
for s in iris['species']:
  if s == 'setosa':
    species.append(1)
  elif s == 'versicolor':
    species.append(2)
  else :
    species.append(3)

print(species)
```

```python
sn.scatterplot(x = 'sepal_length', y = 'petal_length', c = species, data = iris)
plt.show()
```
